```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data():

    file_path = os.path.join('data', 'simulation_results', 'plot_2', 'r.csv')
    data_r = np.loadtxt(file_path, skiprows=1, delimiter=',')

    file_path = os.path.join('data', 'simulation_results', 'plot_2', 'm.csv')
    data_m = np.loadtxt(file_path, skiprows=1, delimiter=',')

    return data_r, data_m

def plot_data(data_r, data_m):
    # ------------------- plot ------------------
    fig, axs = plt.subplots(1, 2, figsize=(10, 4))
   
    axs[0].plot(data_r[:,0], data_r[:,1], label=f'r - grid', linewidth=3, alpha=0.7)
    axs[0].plot(data_m[:,0], data_m[:,1], label=f'm - grid', linestyle='--')
    axs[0].set_xlabel('Radius [cm]')
    axs[0].set_ylabel('Mass [g]')
    axs[0].grid(True, alpha=0.3)
    axs[0].legend()

    axs[1].plot(data_r[:,0], data_r[:,2], label=f'r - grid', linewidth=3, alpha=0.7)
    axs[1].plot(data_m[:,0], data_m[:,2], label=f'm - grid', linestyle='--')
    axs[1].set_yscale('log')
    axs[1].set_xlabel('Radius [cm]')
    axs[1].set_ylabel('Pressure [dyn/cm]')
    axs[1].grid(True, alpha=0.3)
    axs[1].legend()


    plt.suptitle('Jupiter \u2013 Ideal Gas \u2013 RK45')
    
    file_name = 'plot_2_m_vs_r_grid.pdf'

    folder_path = os.path.join('plots')
    file_path = os.path.join(folder_path, file_name)
    os.makedirs(folder_path, exist_ok=True)
    plt.savefig(file_path)

    plt.show()
    plt.close()

data_r, data_m = load_data()
plot_data(data_r, data_m)
```
