```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data():

    file_path = os.path.join('data', 'simulation_results', 'plot_2', 'r.csv')
    data_r = np.loadtxt(file_path, skiprows=1, delimiter=',')

    file_path = os.path.join('data', 'simulation_results', 'plot_2', 'm.csv')
    data_m = np.loadtxt(file_path, skiprows=1, delimiter=',')

    return data_r, data_m

def plot_data(data_r, data_m):
    # ------------------- plot ------------------
    plt.Figure(figsize=(10,4))
    plt.plot(data_r[:,0], data_r[:,1], label=f'r - grid', linewidth=3)
    plt.plot(data_m[:,0], data_m[:,1], label=f'm - grid', linestyle='--')
    # plt.yscale('log')
    plt.xlabel('Radius [cm]')
    plt.ylabel('Mass [g]')
    plt.title('Jupiter \u2013 Polytropic \u2013 RK45')
    plt.grid(True, alpha=0.3)
    plt.legend()

    
    file_name = 'plot_2_m_vs_r_grid.pdf'

    folder_path = os.path.join('plots')
    file_path = os.path.join(folder_path, file_name)
    os.makedirs(folder_path, exist_ok=True)
    plt.savefig(file_path)

    plt.show()
    plt.close()

data_r, data_m = load_data()
plot_data(data_r, data_m)
```
