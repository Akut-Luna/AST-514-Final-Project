```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name = f'Earth_04.5_analytical_core_RK45_theta_2.csv'
file_path = os.path.join('data', 'simulation_results', 'plot_core', file_name)
data = np.loadtxt(file_path, skiprows=1, delimiter=',') 

file_path = os.path.join('data', 'Literature', 'Earth', 'PREM500.csv')
df = pd.read_csv(file_path, names=['radius','density','Vpv','Vsv','Q-kappa','Q-mu','Vph','Vsh','eta'])
lit_radii = df['radius'].to_numpy() * 100 # m -> cm
lit_densities = df['density'].to_numpy() * 1e-3 # kg/m^3 -> g/cm^3
lit_name = 'PREM'
```

```python
plt.figure(figsize=(12,4))
R0 = data[:,0][0]
plt.plot(lit_radii/R0, lit_densities, label='PREM', color='black')
plt.plot(data[:,0]/R0, data[:,3], label='Semi-Analytic MgSiO$_3$ + Fe', linestyle='--', color='red')
plt.title('Earth')
plt.xlabel('Normalised radius')
plt.ylabel('Density [g/cmÂ³]')
plt.grid(True, alpha=0.3)
plt.legend()

plt.savefig('plots/plot_6_core.pdf')
plt.show()
```
